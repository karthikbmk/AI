##Brute Force 9 * 9 Sudoku Solver

```python
import numpy as np
from IPython.display import clear_output
import time
import sys
from IPython.core.debugger import Tracer
sys.setrecursionlimit(10000)
x=0
```

```python
def createFlagBoard(board):
    #Return a board with ones and zeroes.
    # 1 == fixed, 0 == vary
    rows = board.shape[0]
    cols = board.shape[1]
    flag_board = np.zeros((rows,cols))
    
    for i in range(rows):
        for j in range(cols):
            if board[i,j] in range(1,10):
                flag_board[i,j] = 1
    
    return flag_board

def get_pos(actual_board,flag_board,pos_type,cur_pos):
    #Get the previous or next mutable position
    # P = previous, N = next
    
    max_rows = actual_board.shape[0]
    max_cols = actual_board.shape[1]
    
    cur_row = cur_pos[0]
    cur_col = cur_pos[1]
    
    if pos_type == 'P':        

        cur_col -= 1
        last_col = -1
        last_row = -1
        step = -1
        col_reset = max_cols -1
        
    elif pos_type == 'N':
        
        cur_col += 1
        last_col = max_cols
        last_row = max_rows
        step = 1
        col_reset = 0
        
    for i in xrange(cur_row,last_row,step):
        for j in xrange(cur_col,last_col,step):                
            if flag_board[i,j] == 0:
                return (i,j)                
        cur_col = col_reset
    
    #Ideally shouldn't come here
    return (-1,-1)

def canPlace(board,pos,num):
    #Returns True if no constraints violated. else false
    row = pos[0]
    col = pos[1]
    
    #Row check
    if num in board[pos[0],]:
        return False
    
    #Col check
    if num in board[:,pos[1]]:
        return False
    
    #3*3 matrix comparisions            
    #Mat 1,2,3
    if row in range(0,3):
        if col in range(0,3):
            if num in board[0:3,0:3]:
                return False
    
        elif col in range(3,6):
            if num in board[0:3,3:6]:
                return False    
    
        elif col in range(6,9):
            if num in board[0:3,6:9]:
                return False
            
    
    elif row in range(3,6):
        if col in range(0,3):
            if num in board[3:6,0:3]:
                return False
    
        elif col in range(3,6):
            if num in board[3:6,3:6]:
                return False
    
        elif col in range(6,9):
            if num in board[3:6,6:9]:
                return False
    
    elif row in range(6,9):
        if col in range(0,3):
            if num in board[6:9,0:3]:
                return False
    
        elif col in range(3,6):
            if num in board[6:9,3:6]:
                return False   
    
        elif col in range(6,9):
            if num in board[6:9,6:9]:
                return False       
            
    return True

def solve(board,flag_board,pos,num,x):
    
    x += 1
    
    if x%500 == 0:
        print x
    #clear_output()
    #print board
    
    
    if get_pos(board,flag_board,'N',pos) == (-1,-1):        
        for i in range(1,10):
            if i not in board[pos[0],]:
                board[pos[0],pos[1]] = i
                return board
    
    elif canPlace(board,pos,num) == True:
        board[pos[0],pos[1]] = num
        pos = get_pos(board,flag_board,'N',pos)
        num = 1
        
    elif num <= 8:
        num+= 1
    
    else:
        
        if num == 9:
            board[pos[0],pos[1]] = 0
        
        pos = get_pos(board,flag_board,'P',pos)
        num = board[pos[0],pos[1]]
        
        while(num == 9):
            board[pos[0],pos[1]] = 0
            pos = get_pos(board,flag_board,'P',pos)
            num = board[pos[0],pos[1]]
            
        #check here
        num += 1
        
    return solve(board,flag_board,pos,num,x)
```
